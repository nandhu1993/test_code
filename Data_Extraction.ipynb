```
def read_data():
    """ Load the training master data
   
    Returns
    -------
    DataFrame
        Return the DateFrame
    """
    print('Reading the data')
    #------------------------------------------------------------------------------#
    # Read the data
    #------------------------------------------------------------------------------#
    #Sales data
    sales = spark.read.csv(actual+config['Input']['flsales'],low_memory=False)
   
    # 52 weeks calculation
    week_num = spark.read.csv(config['Input']['flcalendar'])
    #Customer File
    store_info = spark.read_excel(config['Input']['flstore_info'])
    # Demographic
    demo = spark.read.csv(config['Input']['fldemo'])
    #Cluster Data
    clus = spark.read.csv(config['Input']['flclus_8'],low_memory=False)    
    return sales,week_num,store_info,demo,clus
```

```
def process_data():
    """ Filter the regions and channel code
    Cast the type of the columns in dataframe
    
    Returns
    -------
    DataFrame
        Return the DateFrame 
    """  
    new_path = '/dbfs/FileStore/innovationforecast/input/'
    sales,week_num,store_info,demo,clus= read_data()
    #------------------------------------------------------------------------------#
    # Data preprocessing 
    #------------------------------------------------------------------------------#
    # Stores info renaming columns
    df_store = store_info.rename({'Cust Nm' : 'Cust_Nm','Cust Nbr':'Cust_Nbr','Chain Lvl1 Node Nm' : 'Chain_Lvl1_Node_Nm',
        'Trade Chanl Code' : 'Trade_Chanl_Code','Loc Nbr' : 'Loc_Nbr','Loc Nm':'Loc_Nm','Geo Lvl 2 Name':'Region',
        'Geo Lvl 3 Name':'Zone','Phy Loc City' : 'Phy_Loc_City','Tkt Sale Amt CY Roll 52 Wk LFW':'Sale_52wk'}, axis=1)
    df_store['Region']= df_store['Region'].str.strip()
    
    upc = config['Input']['flupc']
    flyr = config['Input']['flyr']    
    upc_out = upc
    upc_out=str(upc_out)
    
    # Filtering selected columns
    sales_1 = sales.loc[(sales['prod_upc']== upc) &(sales['fscl_yr_num']== flyr)]
    sales_1 = sales_1[['fl_cust_id','prod_upc','fscl_wk_end_dte','fscl_yr_num','fscl_pd_num','fscl_wk_in_pd_num','net_tkt_sale_amt','net_tkt_sale_qty']]
    sales_cnt = sales_1.count()[0]
    print('Number of Rows in Sales DataFrame :',sales_cnt)
    sales_1.replace('(^\s+|\s+$)', '', regex=True, inplace=True)
    sales_1['net_tkt_sale_amt'] = sales_1.withColumn('net_tkt_sale_amt').cast("integer")
    sales_1['net_tkt_sale_qty'] = sales_1.withColumn('net_tkt_sale_qty').cast("integer")
    
    week_num.fscl_wk_end_dte = spark.to_date(week_num.fscl_wk_end_dte)
    
    # Time of output file generated
    cur_time= datetime.now().strftime("%Y_%m_%d-%I_%M_%p")
    master_file=new_path+'actuals_le_2021_'+ upc_out +'.csv'
    print(':: Master data process started ::')
    #del sales
    #------------------------------------------------------------------------------#
    # Master data preparation 
    #------------------------------------------------------------------------------#
  
    df_cust= spark.createDataFrame()
    df_sales= spark.createDataFrame()
    year_df = week_num.loc[week_num['fscl_yr_num']==flyr]
    sales_df= sales_1.loc[sales_1['fscl_yr_num']==flyr]
    df_innovation = sales_df
    
    df_innovation['net_tkt_sale_qty_postve']=df_innovation['net_tkt_sale_qty']
    df_innovation['net_tkt_sale_qty_neg']=df_innovation['net_tkt_sale_qty']
    df_innovation.loc[df_innovation['net_tkt_sale_qty_postve']<0,'net_tkt_sale_qty_postve']=0
    df_innovation.loc[df_innovation['net_tkt_sale_qty_neg']>0,'net_tkt_sale_qty_neg']=0
    df_innovation['fscl_wk_end_dte'] = spark.to_date(df_innovation['fscl_wk_end_dte'])
    df_innovation['net_tkt_sale_qty_postve'] = df_innovation.withColumn('net_tkt_sale_qty_postve').cast("integer")
    df_innovation['net_tkt_sale_qty_neg'] = df_innovation.withColumn('net_tkt_sale_qty_neg').cast("integer")
	# Shipment Qty needs to be calcualted - in weeks positive only
	# Assume we have only UPC
    df_innovation['Average_waste']=df_innovation.groupby(['fl_cust_id','fscl_yr_num'])['net_tkt_sale_qty_neg'].transform('mean')
    df_innovation['Average_waste']=abs(df_innovation['Average_waste'])
    df_innovation['Average_waste'] = df_innovation['Average_waste'].apply(np.ceil)
    df_innovation['net_tkt_sale_qty_postve_adj']=df_innovation['net_tkt_sale_qty_postve']-df_innovation['Average_waste']
    df_innovation.loc[df_innovation['net_tkt_sale_qty_postve_adj']<0,'net_tkt_sale_qty_postve_adj']=0
    df_innovation.fscl_wk_end_dte = spark.to_date(df_innovation.fscl_wk_end_dte) 
    #df_innovation.to_csv(dbfs_connector+output_layer+'xd_df_before_le_'+upc_out+'_'+str(int(flxdupc))+'.csv')
    print(':: Master data computation completed ::')
	
	# WSA Calculation for 52 weeks
    df_store['Sale_52wk']=df_store['Sale_52wk'].replace('(null)',0)
    df_store['Sale_52wk']=df_store.withColumn('Sale_52wk').cast("integer")
    df_store['WSA']=df_store['Sale_52wk'].apply(lambda x:x/52) 
		
	# Joining with Stores data
    final=df_innovation.merge(df_store, how='left',left_on='fl_cust_id',right_on='Cust_Nbr',suffixes=('', '_y')).reset_index()
    final=final[list(final_clus.columns)+['Cust_Nm','Chain_Lvl1_Node_Nm','Trade_Chanl_Code','Region','Zone','Loc_Nm','Phy_Loc_City','WSA']]
    print(':: Joined with store info ::')
    
    #all_upc_final.drop(['prod_upc'], axis=1, inplace = True)
    final.drop(['FL_CUST_ID'], axis=1, inplace = True)
    final.columns = final.columns.str.lower()
    final=final[['fl_cust_id','cust_nm','chain_lvl1_node_nm','trade_chanl_code','region','zone','loc_nm','phy_loc_city','prod_upc',
                 'fscl_yr_num','fscl_pd_num','fscl_wk_in_pd_num','week_num','fscl_wk_end_dte','actual_fscl_wk_end_dte',
                 'net_tkt_sale_qty','net_tkt_sale_qty_postve','net_tkt_sale_qty_neg','average_waste','net_tkt_sale_qty_postve_adj',
                 'cluster_8','wsa']]
    # Output path - Master Data
    final.to_csv(master_file,index=False)
    print(':: Master file created :' + master_file + ' ::')       
    # capture the processed timing
    end_time= datetime.now().strftime("%Y_%m_%d-%I_%M_%p")
    print(':: Master file completed at :' + end_time + ' ::')
    return final,sales,upc_out

# COMMAND ----------
```

```
def validation():
    new_path = '/dbfs/FileStore/innovationforecast/input/'
    #------------------------------------------------------------------------------#
    #Validation
    #------------------------------------------------------------------------------#
    print(':: Master data validation ::')
    final,sales,upc_out = process_data()
    # Sales data validation
    valid_sales = sales.groupby(['geo_lvl2_node_nm','prod_upc','fscl_yr_num'])['net_tkt_sale_qty'].sum().reset_index()
    valid_sales.to_csv(new_path+'valid_sales_le_act_'+upc_out +'.csv')
    # Processed intermediate data validation    
    valid_final = final.groupby(['region','prod_upc','fscl_yr_num'])['net_tkt_sale_qty','net_tkt_sale_qty_postve',
                                       'net_tkt_sale_qty_neg'].sum().reset_index()
    valid_final.to_csv(new_path+'valid_final_le_act_'+upc_out+'.csv')
    #valid_final.to_csv(dbfs_connector+output_layer+config['Validation']['valid_final'])
    # Masterdata validation
    # Count records - year wise                    
    valid_count=final.groupby(['region','prod_upc','fscl_yr_num']).size().reset_index(name='counts') 
    #valid_count.to_csv(dbfs_connector+output_layer+config['Validation']['valid_count'])
    valid_count.to_csv(new_path+'valid_count_le_act_'+upc_out+'.csv')
    valid_store_count = final.groupby(['region','prod_upc','fl_cust_id','fscl_yr_num'])['net_tkt_sale_qty','net_tkt_sale_qty_postve',
                                       'net_tkt_sale_qty_neg'].sum().reset_index()
    valid_store_count.to_csv(new_path+'valid_store_count_le_act_'+upc_out +'.csv')
    #valid_store_count.to_csv(dbfs_connector+output_layer+config['Validation']['valid_store_count'])

# COMMAND ----------

#  Master data start Processing & Log file
```
